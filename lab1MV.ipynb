```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
from PIL import Image

im = pltimg.imread('6_26.jpg')
cropped_img = im[240:600,890:1200,:]
plt.imsave('cropped_image.jpg', cropped_img)
img = Image.open('6_26.jpg', 'r')
img2 = Image.open('cropped_image.jpg', 'r')
img.paste(img2)
img.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
im = pltimg.imread('6_26.jpg')
fig = plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(im[240:600,890:1200,:])
plt.subplot(122)
plt.imshow(im[0:200,0:200,:], 1500, )
plt.show()
```
